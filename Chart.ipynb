```python
import matplotlib.pyplot as plt
import numpy as np

def create_radial_chart(output_path):
    """
    Creates a radial comparison chart with smoother arcs and a redesigned horizontal-bar legend.

    Args:
        output_path (str): Path to save the output PNG file.
    """
    # Data
    optimized_score = 20  # Optimized score percentage
    now_score = 80        # Current score percentage
    max_score = 100       # Maximum score

    # Normalize scores to angles (in degrees)
    optimized_angle = (optimized_score / max_score) * 360
    now_angle = (now_score / max_score) * 360

    # Create figure and polar axis
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'})
    ax.set_theta_direction(-1)  # Reverse direction for clockwise rotation
    ax.set_theta_offset(np.pi / 2)  # Start from the top (12 o'clock position)

    # Smoother resolution for arcs
    resolution = 1000
    theta_full = np.linspace(0, 2 * np.pi, resolution)

    # Plot the background arcs (full 360 degrees)
    ax.plot(theta_full, [1.2] * len(theta_full), linewidth=30, color='#F0F0F0', zorder=1)  # Outer background
    ax.plot(theta_full, [0.9] * len(theta_full), linewidth=30, color='#F0F0F0', zorder=1)  # Inner background

    # Plot the "Optimized" arc (outer ring)
    theta_optimized = np.linspace(0, np.deg2rad(optimized_angle), resolution)
    ax.plot(theta_optimized, [1.2] * len(theta_optimized), linewidth=30, color='#1E1E4A', zorder=2)

    # Add rounded edges for "Optimized"
    ax.scatter(np.deg2rad(optimized_angle), [1.2], color='#1E1E4A', s=300, zorder=3)  # End of the arc
    ax.scatter(0, [1.2], color='#1E1E4A', s=300, zorder=3)  # Start of the arc

    # Plot the "Now" arc (inner ring)
    theta_now = np.linspace(0, np.deg2rad(now_angle), resolution)
    ax.plot(theta_now, [0.9] * len(theta_now), linewidth=30, color='#F2C94C', zorder=2)

    # Add rounded edges for "Now"
    ax.scatter(np.deg2rad(now_angle), [0.9], color='#F2C94C', s=300, zorder=3)  # End of the arc
    ax.scatter(0, [0.9], color='#F2C94C', s=300, zorder=3)  # Start of the arc

    # Add central text
    ax.text(0, 0, 'Sleep Efficiency', ha='center', va='center', fontsize=18, color='black',
            bbox=dict(facecolor='white', edgecolor='none'))

    # Style adjustments
    ax.set_yticks([])
    ax.set_xticks([])
    ax.set_ylim(0, 1.5)
    ax.spines['polar'].set_visible(False)
    ax.grid(False)

    # Redesigned legend
    legend_y = 0.05  # Position for the legend
    fig.text(0.35, legend_y, '▇', color='#1E1E4A', fontsize=16, va='center')  # Optimized bar
    fig.text(0.39, legend_y, 'Optimised', ha='left', va='center', fontsize=12, color='black')

    fig.text(0.56, legend_y, '▇', color='#F2C94C', fontsize=16, va='center')  # Now bar
    fig.text(0.6, legend_y, 'Now', ha='left', va='center', fontsize=12, color='black')

    # Add title below the chart
    plt.title("SLEEP EFFICIENCY", va='top', fontsize=16, fontweight='bold')

    # Save the chart
    fig.savefig(output_path, format="png", dpi=300, bbox_inches="tight")
    #plt.close(fig)

# Example usage
output_path = "/Users/rajathtalpady/Desktop/radial_comparison_chart_fixed_legend.png"
create_radial_chart(output_path)
```

```python
import matplotlib.pyplot as plt

# Data for the donut chart
optimised_value = 75  # Percentage value for Optimised
now_value = 25        # Percentage value for Now

# Plot settings
outer_colors = ['#1E1F3F', 'lightgrey']  # Dark navy blue for Optimised and grey for rest
inner_colors = ['#F1C93B', 'lightgrey']  # Yellow for Now and grey for rest

# Create figure
fig, ax = plt.subplots(figsize=(6, 6))

# Plot outer ring (Optimised)
wedges_outer, _ = ax.pie([optimised_value, 100-optimised_value], 
                         radius=1, 
                         colors=outer_colors, 
                         startangle=90,
                         counterclock=False, 
                         wedgeprops=dict(width=0.27, edgecolor='w'))

# Plot inner ring (Now)
wedges_inner, _ = ax.pie([now_value, 100-now_value], 
                         radius=0.7, 
                         colors=inner_colors, 
                         startangle=90,
                         counterclock=False,  
                         wedgeprops=dict(width=0.27, edgecolor='w'))

# Add the central text
plt.text(0, 0, 'Sleep Efficiency', ha='center', va='center', fontsize=12, color='gray')

# Add the legend with correct handles and adjusted key size
plt.legend([wedges_outer[0], wedges_inner[0]],  # Correct handles
           ['Optimised', 'Now'], 
           loc="upper center", 
           bbox_to_anchor=(0.5, -0.1), 
           ncol=2,
           handlelength=2,  # Adjust the horizontal length of the key bar
           handleheight=0.2,  # Adjust the vertical size
           fontsize=10)  # Adjust legend font size

# Title for the chart
plt.title("SLEEP EFFICIENCY", fontsize=20, weight='bold', pad= 20)

# Ensure the pie is drawn as a circle
ax.axis('equal')  

# Display the chart
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors

def create_radial_chart(
    csv_path, 
    output_path, 
    optimized_color="blue",  # Default to color names, e.g., "blue"
    now_color="orange",      # Default to color names, e.g., "orange"
    central_text="Sleep Efficiency", 
    title="SLEEP EFFICIENCY"
):
    """
    Creates a radial comparison chart with customizable colors, text, and data from a CSV.

    Args:
        csv_path (str): Path to the CSV file containing the scores.
        output_path (str): Path to save the output PNG file.
        optimized_color (str): Color for the optimized arc (can be a color name or hex code).
        now_color (str): Color for the now arc (can be a color name or hex code).
        central_text (str): Text to display in the center of the chart.
        title (str): Title displayed below the chart.
    """
    # Convert color names to hex codes
    optimized_color = colors.to_rgba(optimized_color, 1)  # Convert to RGBA format
    now_color = colors.to_rgba(now_color, 1)              # Convert to RGBA format

    # Load data from the CSV file
    data = pd.read_csv(csv_path)
    optimized_score = data["Optimized Score"].iloc[0]
    now_score = data["Now Score"].iloc[0]
    max_score = 100  # Assuming the scores are percentages

    # Normalize scores to angles (in degrees)
    optimized_angle = (optimized_score / max_score) * 360
    now_angle = (now_score / max_score) * 360

    # Create figure and polar axis
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'})
    ax.set_theta_direction(-1)
    ax.set_theta_offset(np.pi / 2)

    # Plot full background arcs
    resolution = 1000
    theta_full = np.linspace(0, 2 * np.pi, resolution)
    ax.plot(theta_full, [1.2] * len(theta_full), linewidth=30, color="#F0F0F0", zorder=1)
    ax.plot(theta_full, [0.9] * len(theta_full), linewidth=30, color="#F0F0F0", zorder=1)

    # Plot optimized arc
    theta_optimized = np.linspace(0, np.deg2rad(optimized_angle), resolution)
    ax.plot(theta_optimized, [1.2] * len(theta_optimized), linewidth=30, color=optimized_color, zorder=2)
    ax.scatter(np.deg2rad(optimized_angle), [1.2], color=optimized_color, s=300, zorder=3)
    ax.scatter(0, [1.2], color=optimized_color, s=300, zorder=3)

    # Plot now arc
    theta_now = np.linspace(0, np.deg2rad(now_angle), resolution)
    ax.plot(theta_now, [0.9] * len(theta_now), linewidth=30, color=now_color, zorder=2)
    ax.scatter(np.deg2rad(now_angle), [0.9], color=now_color, s=300, zorder=3)
    ax.scatter(0, [0.9], color=now_color, s=300, zorder=3)

    # Add central text
    ax.text(0, 0, central_text, ha="center", va="center", fontsize=18, color="black",
            bbox=dict(facecolor="white", edgecolor="none"))

    # Style adjustments
    ax.set_yticks([])
    ax.set_xticks([])
    ax.set_ylim(0, 1.5)
    ax.spines["polar"].set_visible(False)
    ax.grid(False)

    # Add legend
    fig.text(0.35, 0.05, "▇", color=optimized_color, fontsize=16, va="center")
    fig.text(0.39, 0.05, "Optimized", ha="left", va="center", fontsize=12, color="black")
    fig.text(0.56, 0.05, "▇", color=now_color, fontsize=16, va="center")
    fig.text(0.6, 0.05, "Now", ha="left", va="center", fontsize=12, color="black")

    # Add dynamic title
    plt.title(title, va="top", fontsize=16, fontweight="bold")

    # Save the chart
    fig.savefig(output_path, format="png", dpi=300, bbox_inches="tight")
```

```python
import argparse
from radial_chart import create_radial_chart  # Ensure this imports your updated function

def main():
    # Set up argument parsing
    parser = argparse.ArgumentParser(description="Generate a radial comparison chart.")
    parser.add_argument("csv_path", type=str, help="Path to the CSV file containing the scores.")
    parser.add_argument("output_path", type=str, help="Path to save the generated chart.")
    parser.add_argument("--optimized_color", type=str, default="#1E1E4A", help="Color for the optimized arc.")
    parser.add_argument("--now_color", type=str, default="#F2C94C", help="Color for the now arc.")
    parser.add_argument("--central_text", type=str, default="Sleep Efficiency", help="Central text in the chart.")
    parser.add_argument("--title", type=str, default="SLEEP EFFICIENCY", help="Title displayed below the chart.")

    # Parse command-line arguments
    args = parser.parse_args()

    # Pass arguments to the create_radial_chart function
    create_radial_chart(
        csv_path=args.csv_path,
        output_path=args.output_path,
        optimized_color=args.optimized_color,
        now_color=args.now_color,
        central_text=args.central_text,
        title=args.title
    )

    print(f"Radial chart saved at: {args.output_path}")

if __name__ == "__main__":
    main()
```
